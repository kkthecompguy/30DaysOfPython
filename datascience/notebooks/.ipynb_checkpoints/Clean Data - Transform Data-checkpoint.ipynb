```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data') 
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
os.makedirs(CACHE_DIR, exist_ok=True)
```

```python
csvfiles = [x for x in os.listdir(DATA_DIR) if x.endswith('.csv')] 
my_dataframe = []
for filename in csvfiles:
    year = filename.replace('.csv', '')
    df = pd.read_csv(os.path.join(DATA_DIR, filename))
    df['Filename'] = filename
    df['Year'] = year 
    my_dataframe.append(df)
```

```python
 
my_entire_dataframe = pd.concat(my_dataframe)
my_entire_dataframe.head()
```

```python
my_entire_dataframe.tail()
```

```python
output = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
my_entire_dataframe.to_csv(output, index=False)
```
