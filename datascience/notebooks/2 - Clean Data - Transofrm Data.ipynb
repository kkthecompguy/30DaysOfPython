```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'movies')
CACHE = os.path.join(BASE_DIR, 'cache')
dataset = os.path.join(CACHE, 'themoviedb-search.csv') 
output_file = os.path.join(CACHE, 'themoviedb-dataset-cleaned.csv')
```

```python
df = pd.read_csv(dataset)
df.head()
```

```python
df.drop(['backdrop_path', 'belongs_to_collection', 'genres', 'homepage', 'original_title', 'runtime', 'spoken_languages', 'poster_path', 'production_companies', 'production_countries'], axis=1, inplace=True)
df.head()
```

```python
df = df.sort_values(by='popularity', ascending=False)
df.head()
```

```python
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df.to_csv(output_file, index=False)
```
