```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data') 
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
dataset = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
output_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset-cleaned.csv')
```

```python
df = pd.read_csv(dataset)
df.head()
```

```python
df['Domestic %'] = df['%']
df['Foreign %'] = df['%.1']
df.head()
```

```python
df.drop(['%', '%.1'], axis=1, inplace=True)
df.head()
```

```python
to_clean_columns = ['Worldwide', 'Domestic', 'Foreign']

def currency_str_to_int(current_val):
    currency_val = current_val.replace("$","").replace(",","")
    try:
        currency_val = int(currency_val)
    except:
        currency_val = 0
    return currency_val

def clean_col(row):
    for col in to_clean_columns:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row

df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = -1
df_cleaned.head()
```

```python
df_cleaned.sort_values(by='Worldwide', inplace=True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index + 1
df_cleaned.head() 
```

```python
df_cleaned['Domestic %'] = (df_cleaned['Domestic'] / df_cleaned['Worldwide']) * 100
df_cleaned['Foreign %'] = (df_cleaned['Foreign'] / df_cleaned['Worldwide']) * 100
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.to_csv(output_file, index=False)
```
